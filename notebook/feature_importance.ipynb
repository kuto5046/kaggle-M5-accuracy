```python
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import lightgbm as lgb
```

```python
with open('../model/model_20200405_001355.picklemodel.pickle', mode='rb') as fp:
    model = pickle.load(fp)
```

```python
fig, ax = plt.subplots(figsize=(10,16), tight_layout=True)
sns.set_palette("Blues_r")
lgb.plot_importance(model, ax=ax, height=0.2, xlim=None, ylim=None, title='Feature importance', 
                    xlabel='Importance', ylabel='Features', importance_type='gain', max_num_features=None, 
                    ignore_zero=True, figsize=None, dpi=None, grid=True, precision=3)
```
